```python
#Import libraries

import pandas as pd
import os
```

```python
#Filepath
filename = os.path.join(os.getcwd(),"Customer_Rating.csv")

try:
    #Open and store file contents to dataframe
    data = pd.read_csv(filename)
except:
    print("File not found.")
                        
```

```python
#1 - Show the last five records
num_rows = len(data)  #Total number of rows

data[num_rows - 5:]    #Get last 5 records
```

```python
#2- Data of customers who continued their subscription
d2 = data[data["Continue?"]=="Yes"]

d2
```

```python
#3 - Category, address, and sales data of customers who continue their subscription

d3 = data[data["Continue?"] == "Yes"]     #customers who continue their subscription
d3 = d3[["Category","Address","Sales"]]   #select category, address, and sales columns

d3
```

```python
#4 - Data of customers who continued their subscription with sales less than 70000

d4 = data[(data["Continue?"]== "Yes")]   #Customers who continue their subscription
d4 = d4[d4["Sales"] < 70000]             # Filter customers with sales < 70000 

d4
```

```python
#5 - Write data in item 4 to empty.csv

#Name of outputfile
file_out = os.path.join(os.getcwd(),"empty.csv")

try:
    d4.to_csv(file_out, index = False)
    print("Please see output file: " + file_out)
except:
    print("Error writing file")
    
```
