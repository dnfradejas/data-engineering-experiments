```python
import pandas as pd

# read sales.csv
data = pd.read_csv("df1.csv")
data1 = pd.read_csv("df2.csv")
data2 = pd.read_csv("df3.csv")

# pd concat to merge the csv
x = pd.concat([data,data1,data2])

# save to csv
x.to_csv("df4.csv", index=False)
data3 = pd.read_csv("df4.csv")
print(data3)
```
