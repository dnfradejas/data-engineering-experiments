# Given an initial characteriazation of the entire dataframe

```python
import pandas as pd
sales = pd.read_csv("Fradejas_sales.csv")
sales.describe(include="all")
```

# Remove duplicate row values.

```python
sales = sales.drop_duplicates()
sales
```

# Express price in thousand.

```python
# Remove comma from the column
sales = sales.replace(',','', regex=True)

#remove rows in price column with empty cell
sales = sales[sales.Price.notna()]

# convert data frame to integer
sales['Price'] = sales['Price'].astype(float)

sales.Price = sales.Price / 1000
```

```python
sales.Price
```

# Select name, city, state, and country data of product 2 bought by customers from united states and paid thru Visa or Mastercard. Store the selected data in empty.csv file. Please download the empty.csv file

```python
import pandas as pd

info = pd.read_csv("Fradejas_sales.csv")
```

```python
# filter results by condition
info = info[(info.Product == 'Product2') & (info.Country == 'United States') & ((info.Payment_Type == 'Visa') | ( info.Payment_Type=='Mastercard'))]

# select data needed after selection
new_info = info[['Name', 'City', 'State', 'Country', 'Latitude', 'Longitude']]

# writing to a file
new_info.to_csv('empty.csv', index=False)
```

```python
data = pd.read_csv("Fradejas_empty.csv")
print(data)
```

# Select name, price, city, state and country data of product 2 bought by customers from New York paid thru Mastercard. Store the selected data in empty1.csv file. Please download the empty1.csv file.

```python
import pandas as pd

info = pd.read_csv("Fradejas_sales.csv")

# remove trailing spaces 
info['City'] = info['City'].str.strip()

# filter results by condition
info = info[(info.Product=='Product2') & ((info.City=='New York') & (info.Payment_Type=='Mastercard'))]
info
```

```python
# select data needed after selection
new_info = info[['Name', 'Price', 'City', 'State', 'Country', 'Latitude', 'Longitude']]
new_info

# writing to a file
new_info.to_csv('empty1.csv', index=False)
```

```python
data = pd.read_csv("Fradejas_empty1.csv")
print(data)
```
