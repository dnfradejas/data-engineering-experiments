```python
import requests
import bs4
import lxml

req = requests.get('https://en.wikipedia.org/wiki/Python_(programming_language)') # requests object
req.text # html code
data = bs4.BeautifulSoup(req.text, 'lxml') # beautiful soup object

Title = data.select('title') #selecting text with tag title
Title[0].getText() # extracting text only
```

```python
head = data.select('.mw-headline') # selecting text with tag .mw headline
x = len(head)
for row in range(x):
    print(head[row].getText())
```

```python
# importing libraries
from bs4 import BeautifulSoup
import urllib.request
import re
```

```python
url = "https://en.wikipedia.org/wiki/Python_(programming_language)"
page = urllib.request.urlopen(url) #connect to website
soup = BeautifulSoup(page, 'html.parser')
```

```python
contents = soup.find_all('p')
print(contents)
```

```python
content = []
for li in contents:
    content.append(li.getText().split('\n')[0])
print(content)
```
