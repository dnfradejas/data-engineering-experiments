```python
#import necessary libraries
import pandas as pd
import numpy as np
```

```python
#import the three csv files

hotels = pd.read_csv('Hotel.csv')
restaurants = pd.read_csv('Restaurant.csv', index_col = False)
tourist_sites = pd.read_csv('Tourist_site.csv', index_col = False)
```

```python
#add a new column indicating the location type

hotels['Type'] = 'Hotel'
restaurants['Type'] = 'Restaurant'
tourist_sites['Type'] = 'Tourist Site'
```

```python
#standardize the column names to align it with the other datasets
hotels = hotels.rename(columns = {'hotel_name':'name'}, inplace = False)

#drop index column of hotels ('Unnamed: 0')
hotels = hotels.drop(columns = 'Unnamed: 0')
```

```python
#Create a list of metro manila cities
cities = [', Caloocan',
    ', Las Pinas',
    ', Makati',
    ', Malabon',
    ', Mandaluyong',
    ', Manila',
    ', Marikina',
    ', Muntinlupa',
    ', Navotas',
    ', Paranaque',
    ', Pasay',
    ', Pasig',
    ', Quezon City',
    ', San Juan',
    ', Taguig',
    ', Valenzuela'
]

#add City column on each dataset
hotels['City'] = ''
restaurants['City'] = ''
tourist_sites['City'] = ''
```

```python
#get the size of the hotel dataframe
hotels_shape = hotels.shape

#Using for loop, assign values to the City column of the hotel dataset
for i in range(0, len(cities)):
    for j in range(0,len(hotels)):
        #if an element of cities is present in the address element of hotel, put the city as value of the city column of the hotel
        if cities[i] in hotels.iloc[j,2]: 
            hotels.iloc[j,hotels_shape[1]-1] = cities[i].replace(',','')
            
#sort the values of hotels by City
hotels = hotels.sort_values(by = 'City', ascending = True)
```

```python
#check content
hotels
```

```python
#standardize the column names to align it with the other datasets
restaurants = restaurants.rename(columns = {'restaurant_name':'name', 'price_range':'price'})

#drop index column of restaurants ('Unnamed: 0')
restaurants = restaurants.drop(columns = ['Unnamed: 0'])

#get the size of the restaurants dataframe
restaurants_shape = restaurants.shape
```

```python
#Using for loop, assign values to the City column of the restaurant dataset
for i in range(0, len(cities)):
    for j in range(0,len(restaurants)):
        #if an element of cities is present in the address element of hotel, put the city as value of the city column of the restaurant
        if cities[i] in restaurants.iloc[j,2]:
            restaurants.iloc[j,restaurants_shape[1]-1] = cities[i].replace(',','')
            
#sort values of restaurant dataset by City
restaurants = restaurants.sort_values(by = 'City', ascending = True)
```

```python
#check content
restaurants
```

```python
#standardize the column names to align it with the other datasets
tourist_sites = tourist_sites.rename(columns = {'Tourist Site':'name','Entrance Fee (PHP)':'price'})

#drop index column of tourist_sites ('Unnamed: 0')
tourist_sites = tourist_sites.drop(columns = ['Unnamed: 0'])

#get the size of the tourist_sites dataframe
tourist_shape = tourist_sites.shape
```

```python
#Using for loop, assign values to the City column of the restaurant dataset
for i in range(0, len(cities)):
    for j in range(0,len(tourist_sites)):
        if cities[i] in tourist_sites.iloc[j,4]:
            tourist_sites.iloc[j,tourist_shape[1]-1] = cities[i].replace(',','')
            
#sort values of tourist_sites dataframe by City
tourist_sites = tourist_sites.sort_values(by = 'City', ascending = True)
```

```python
#check content
tourist_sites
```

```python
#concatenate the three datasets
df_location = pd.concat([hotels,restaurants,tourist_sites])
```

```python
#Put N/A to the columns not applicable to the other datasets using for loop

l = 0
for i in range(0,len(df_location)):
    if ((df_location.iloc[l,36] == 'Restaurant') or (df_location.iloc[l,36] == 'Tourist Site')):
        df_location.iloc[l,4] = 'N/A' #no of rooms N/A to Restaurant and Tourist Site
        df_location.iloc[l,5] = 'N/A' #amenities N/A to Restaurant and Tourist Site
    if ((df_location.iloc[l,36] == 'Hotel') or (df_location.iloc[l,36] == 'Tourist Site')):
        df_location.iloc[l,38] = 'N/A' #specialty N/A to Hotel and Tourist Site
        df_location.iloc[l,39] = 'N/A' #special_diets N/A to Hotel and Tourist Site
        df_location.iloc[l,40] = 'N/A' #meals N/A to Hotel and Tourist Site
    if ((df_location.iloc[l,36] == 'Hotel') or (df_location.iloc[l,36] == 'Restaurant')):
        df_location.iloc[l,51] = 'N/A' #classification N/A to Hotel and Restaurant
    if l < len(df_location):
        l = l + 1
```

```python
#remove rows with missing name values
df_location = df_location[df_location['name'] != 'No Data']
```

```python
#Re-order columns

ordered_columns = [
    'City','Type', 'name', 'about', 'Classification', 'address', 'price', 'no_of_rooms', 'amenities','specialty',
    'special_diets','meals','review_1','reviewer_1','review_2','reviewer_2','review_3','reviewer_3','review_4','reviewer_4',
    'review_5','reviewer_5','review_6','reviewer_6','review_7','reviewer_7','review_8','reviewer_8','review_9','reviewer_9',
    'review_10','reviewer_10','review_11','reviewer_11','review_12','reviewer_12','review_13','reviewer_13','review_14',
    'reviewer_14','review_15','reviewer_15','review_16','reviewer_16','review_17','reviewer_17','review_18','reviewer_18',
    'review_19','reviewer_19','review_20','reviewer_20'
]

df_location = df_location[ordered_columns]
```

```python
#remove rows with missing city values
df_location = df_location[df_location['City'] != '']
```

```python
#sort values
df_location = df_location.sort_values(by = ['City','Type','name'], ascending = True)
df_location = df_location.fillna('No Data')

#check content
df_location
```

```python
#export csv files. Data is exported per location
for i in range(0,len(cities)-1):
    df_location[df_location['City'] == cities[i].replace(',','')].to_csv(cities[i].replace(', ','') + '.csv', index = False)
    #df_city.to_csv(i.replace(', ','') + '.csv')
```

```python
#export complete dataset
df_location.to_csv('Location.csv', index = False)
```
