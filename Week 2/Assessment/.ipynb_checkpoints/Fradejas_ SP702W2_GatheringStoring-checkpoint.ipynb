# Optional: Store the data extracted in item 4 in a database (i.e. SQL database).

Sqlite 3 has been used to avoid sharing of database passwords in PostgreSQL

```python
from sqlalchemy import create_engine
import pandas as pd

# read sales.csv
data = pd.read_csv("customer.csv")
data = data[(data.Buy_Car=="Yes") & (data.Income > 30000)]

# Establish Connection
engine = create_engine('sqlite:///customer.db')
sqlite_connection = engine.connect()

# Create Table
sqlite_table = "customer"
data.to_sql(sqlite_table, sqlite_connection, if_exists="replace")

# Close connection
sqlite_connection.close()
```

```python
import sqlite3
db_file = "customer.db"
try:
    # establish connection
    conn = sqlite3.connect(db_file)
    cur = conn.cursor()
    
    # execute query from the table
    cur.execute("Select* From customer")
    rows = cur.fetchall()
    
    # print results
    print(rows) 
    
#throw an exception if there is an error
except Error as e:
    print(e)

    
```
