# Outlier Handling

```python
import numpy as np
import pandas as pd
```

```python
nwdata = pd.read_csv('networth.csv')
newdata.describe(include="all")
```

```python
m = np.mean(nwdata.net_worth)
s = np.std(nwdata.net_worth)

print(m)
print(s)

ll = max(0, m-3*s)
ul = m + 3*s

print(ll)
print(ul)

nwdata_cleaned1 = nwdata[nwdata.net_worth >= ll]
nwdata_cleaned2 = nwdata[nwdata.net_worth <= ul]
```

```python
nwdata_cleaned2.describe(include="all")
```

```python
IQR = np.percentile(nwdata.net_worth, 75) - np.percentile(nwdata.net_worth, 25)
print(IQR)

ul2 = np.percentile(nwdata.net_worth, 75) + 1.5*IQR
ll2 = max(0,np.percentile(nwdata.net_worth, 25) - 1.5*IQR)

print(ul2)
print(ll2)
```

```python
nwdata_cleaned3 = nwdata[nwdata.net_worth >= ll2]
nwdata_cleaned4 = nwdata[nwdata.net_worth <= ul2]
nwdata_cleaned4.describe(include="all")
```
