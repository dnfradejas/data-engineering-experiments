# Removing Duplicate Rows

```python
import pandas as pd
```

```python
data = {"Name": ["Angelo", "Regine", "Raymond", "Angelo"], 
        "Age": [24, 28, 40, 24],
        "Sex": ["Male", "Female", "Male", "Male"]}

profile = pd.DataFrame(data)
print(profile)
```

```python
profile = profile.drop_duplicates()
```

```python
print(profile)
```

```python
profile = profile.drop_duplicates(subset='Sex', keep='first')
```

```python
print(profile)
```
