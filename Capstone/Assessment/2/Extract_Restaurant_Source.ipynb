```python
#import important libraries

import numpy as np
import pandas as pd
import requests
import bs4
from bs4 import BeautifulSoup
import urllib.request
import re #regular expression
import lxml
import json
from time import sleep
from random import randint
```

```python
#get data from Trip Advisor

#Metro Manila Restaurant

url = 'https://www.tripadvisor.com.ph/Restaurants-g298449-Metro_Manila_Luzon.html'

#connect to website
page = urllib.request.urlopen(url) 

#convert to beautiful soup object
soup = BeautifulSoup(page, 'lxml')
print(soup)
```

```python
#create a list where the link of each metro manila restaurants from TripAdvisor will be stored

links_mm = []
```

```python
#find all links of each metro manila restaurant pages from Trip Advisor and store to the links_mm list

for review in soup.findAll('a',{'class':'_15_ydu6b'}):
  a = review['href']
  a = 'https://www.tripadvisor.in'+ a
  a = a[:(a.find('Reviews')+7)] + '-or{}' + a[(a.find('Reviews')+7):]
  links_mm.append(a)
```

```python
#check contents of links
links_mm
```

```python
#This section extracts the restaurant reviews and address from the restaurant pages of Trip Advisor. Note that each page only contains 10 
#reviews and need to click next page to view the succeeding reviews.

#create a list which will contain numbers that will be used to extract reviews from page 1 to page 2 of each restaurant page
d = ['','10']

#create a list which will contain the reviews
reviews = []

#using for loop, extract reviews fron each hotel pages of Travel Advisor
for link in links_mm:
    print(link)
    for i in range(1,len(d)+1):
        
        #replace the {} with the values of d. " ",10 refers to page 1 and 2 of the restaurant page respectively
        
        link2 = link.replace("{}",str(d[i-1]))
        soup2 = urllib.request.urlopen(link2)
        sleep(randint(1,5))
        bsobj2 = BeautifulSoup(soup2,'lxml')
        if i == 1:
            
            #extract the headers, which contains the name of the restaurant and append to reviews
            for s in bsobj2.find_all('h1'):
                reviews.append(s.getText().split('\n')[0])
                
            #extract the restaurant address and append to reviews
            for t in bsobj2.findAll('a','_2wKz--mA _27M8V6YV'):
                reviews.append(t.getText().split('\n')[0])

        #extract reviews and append to reviews
        for v in bsobj2.findAll('div','quote'):
            reviews.append(v.getText().split('\n')[0])
            
        #extract reviewers and append to reviews
            
        for u in bsobj2.findAll('div','ui_column is-2 memberInfoColumn'):
            reviews.append(u.getText().split('\n')[0])
```

```python
#convert reviews to a 30x1 dataframe
columns_1 = ['content']
df =  pd.DataFrame(reviews,columns = columns_1)


#create a second dataframe segregating each restaurant per row and will contain the restaurant name, address, and reviews
columns_2 = ['restaurant_name','restaurant_name2','address','website','review_1','review_2',
              'review_3','review_4','review_5','review_6',
              'review_7','review_8','review_9','review_10',
               'reviewer_1','reviewer_2','reviewer_3','reviewer_4','reviewer_5','reviewer_6','reviewer_7','reviewer_8',
               'reviewer_9','reviewer_10',
              'review_11','review_12','review_13','review_14',
              'review_15','review_16','review_17','review_18',
              'review_19','review_20',
              'reviewer_11','reviewer_12','reviewer_13','reviewer_14','reviewer_15','reviewer_16','reviewer_17','reviewer_18',
              'reviewer_19','reviewer_20']
df_reviews = pd.DataFrame(columns = columns_2, index = range(0,len(links_mm)))

#check structure
df_reviews
```

```python
#transfer the content of df to df_reviews

k = 0
for i in range(1,len(links_mm)):
    for j in range(1,len(columns_2)+1):
        df_reviews.iloc[i-1,j-1] = df.iloc[k,0]
        k=k+1

#check content of df_reviews
df_reviews
```

```python
#This section extracts the price range, cuisines, and specialty of each restaurant. 

#create a list which will contain the extracted data
reviews_2 = []

#using for loop, extract the price range, cuisines, and specialty of each restaurant
for link in links_mm:
    print(link)
        
    #replace the {} with the " ", which refers to the first page of each restaurant pages.
        
    link2 = link.replace("{}",str(''))
    soup2 = urllib.request.urlopen(link2)
    sleep(randint(1,5))
    bsobj2 = BeautifulSoup(soup2,'lxml')
            
    #extract the headers, which contains the name of the restaurant and append to reviews
    for s in bsobj2.find_all('h1'):
        reviews_2.append(s.getText().split('\n')[0])
            
    #extract price range, cuisines, and specialty
    for u in bsobj2.findAll('div','_1XLfiSsv'):
        reviews_2.append(u.getText().split('\n')[0])
```

```python
#create a dataframe which will segregate the price range, specialty, special_diets, and meals per restaurant

df_price_specialty = pd.DataFrame(columns = ['restaurant_name','restaurant_name2','price_range','specialty','special_diets','meals'],index = range(0,len(links_mm)))
df_price_specialty
```

```python
#the price range, specialty, special_diets, and meals have the same html tags. To segregate the attributes, we created a list
#containing the possible values for specialty, and meals

specialty = ['Chinese','Asian','Healthy','Soups','Wine Bar','International','Philippine','Café','Spanish','Bar',
                'Contemporary','Philippine','Fusion','European','American','Steakhouse','Southwestern','Pub','Seafood',
                'Japanense','Sushi','Korean']

meals = ['Breakfast','Lunch','Dinner','Brunch','After-hours','Drinks']
```

```python
#From the reviews_2 list, segregate the restaurant attributes per row using for loop

l = 0
t = 0
u = 0
for i in range(0,28):
    for j in range(0,5):
        if ((l <= len(reviews_2)) & (i != len(links_mm)) &(reviews_2[l] != df_reviews.iloc[i+1,0])):
            for m in range(0,len(specialty)):
                if(specialty[m] in reviews_2[l]):
                    t = t + 1
            for n in range(0,len(meals)):
                if(meals[n] in reviews_2[l]):
                    u = u + 1
            if(reviews_2[l] == df_reviews.iloc[i,0] ):
                df_price_specialty.iloc[i,0] = reviews_2[l]
            elif(reviews_2[l] == df_reviews.iloc[i,1]):
                df_price_specialty.iloc[i,1] = reviews_2[l]
            elif('PHP' in reviews_2[l]):
                df_price_specialty.iloc[i,2] = reviews_2[l]
            elif(t >= 1):
                df_price_specialty.iloc[i,3] = reviews_2[l]
            elif(u >=1):
                df_price_specialty.iloc[i,5] = reviews_2[l]
            elif(j>3):
                df_price_specialty.iloc[i,j] = reviews_2[l]
            l = l+1
            t = 0
            u = 0
        elif (l == len(reviews_2)):
            df_price_specialty.iloc[i,j] = reviews_2[l]
            l = l+1
        else:
            break
        
```

```python
#check the content of df_price_specialty
df_price_specialty
```

```python
#create a dataframe which will segregate the price range, specialty, special_diets, and meals per restaurant

columns_about = ['restaurant_name','restaurant_name2','about']
df_about = pd.DataFrame(columns = columns_about,index = range(0,len(links_mm)))
df_about
```

```python
#This section extracts the price range, cuisines, and specialty of each restaurant. 

#create a list which will contain the extracted data

l = 0
#using for loop, extract the price range, cuisines, and specialty of each restaurant
for link in links_mm:
    print(link)
    about = [] 
    #replace the {} with the " ", which refers to the first page of each restaurant pages.
        
    link2 = link.replace("{}",str(''))
    soup2 = urllib.request.urlopen(link2)
    sleep(randint(1,5))
    bsobj2 = BeautifulSoup(soup2,'lxml')
            
    #extract the headers, which contains the name of the restaurant and append to reviews
    for s in bsobj2.find_all('h1'):
        about.append(s.getText().split('\n')[0])
            
    #extract price range, cuisines, and specialty
    for u in bsobj2.findAll('div', {'class':'_3tRlph6J'}):
        about.append(u.getText().split('\n')[0])
    
    for i in range(0,len(about)):
        if i < len(columns_about):
            df_about.iloc[l,i] = about[i]
    
    if l < len(links_mm)-1:
        l = l +1
```

```python
df_about = df_about.fillna('No Data')
df_about
```

```python
#merge df_restaurant with df_price_specialty

df_restaurant = df_reviews.merge(df_price_specialty, how = 'left', left_on = 'restaurant_name', right_on = 'restaurant_name', suffixes = ['','_specialty'])

#merge df_restaurant with df_about

df_restaurant = df_restaurant.merge(df_about, how = 'left')

#fill na with 'No Data' to indicate that there is no available data from Travel Advisor website
df_restaurant = df_restaurant.fillna('No Data')

#check content
df_restaurant
```

```python
#re-arrange the columns of df_restaurant
df_restaurant = df_restaurant[['restaurant_name','about','address','price_range','specialty','special_diets','meals','review_1','reviewer_1',
              'review_2','reviewer_2','review_3','reviewer_3','review_4','reviewer_4','review_5','reviewer_5','review_6',
              'reviewer_6','review_7','reviewer_7','review_8','reviewer_8','review_9','reviewer_9','review_10','reviewer_10',
              'review_11','reviewer_11','review_12','reviewer_12','review_13','reviewer_13','review_14','reviewer_14',
              'review_15','reviewer_15','review_16','reviewer_16','review_17','reviewer_17','review_18','reviewer_18',
              'review_19','reviewer_19','review_20','reviewer_20']]
df_restaurant
```
