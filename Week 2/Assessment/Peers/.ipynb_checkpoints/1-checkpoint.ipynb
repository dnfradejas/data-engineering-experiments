# PGA 02 - SET 3 - NEW FRANCHISES DATA

```python
import pandas as pd #import pandas library for this PGA
```

## Read CSV Data first then display all data.

```python
data = pd.read_csv('New_Franchise.csv') ## read New_Franchise.csv file to "data"
data ## print whole data to notebook
```

## Item No. 1: The first seven records from the New Franchises.csv File.

```python
data.iloc[0:7] #print first seven records of the dataframe.
```

## Item No. 2: Data of Location where the new franchise was built (Build_new_franchise = Yes)

```python
data[data['Build_New_Franchise']=='Yes'][['ID','Location','Build_New_Franchise']] ##print all records with new franchises
```

## Item No. 3: Average_income_family, Minimum_salary, and Fast_internet data of locations where a new franchise was built

```python
#print the following data with new frnachises given the specified attributes (columns)
data[data['Build_New_Franchise']=='Yes'][['ID','Location','Average_Income_Family','Minimum_Salary','Fast_Internet','Build_New_Franchise']]
```

## Item No. 4: Data of locations  where a new franchise was built and with average family income greater than 27000 

```python
# print the new franchise data given average family income threshold of 27,000.00
data2 = data[(data['Build_New_Franchise']=='Yes') & (data['Average_Income_Family']>27000)][['ID','Location','Average_Income_Family','Build_New_Franchise']]
data2 ## print data2 dataframe
```

## Item No. 5: Store data extracted in item no. 4 in empty.csv file

```python
data2.to_csv('empty.csv',index = False) ## write Item No. 4 Data (Data2) to empty.csv file

#This below is a test if the file writing worked
readfile = pd.read_csv('empty.csv') ## read empty.csv file to variable readfile.
readfile ## Print empt.csv data to check if Item No. 4 Data had been written to file.
```
