```python
#Part 1: Using data found in sales.csv. Perform the following task:
# 1. Standardize the Price column
# 2. Identify and remove outliers in the dataset. 


sales = pd.read_csv('sales.csv')

#Standardiztion of Price
sales.insert(3, "standardized_nw", 0)
sales.Price = sales.Price.replace('[,]', '', regex=True).astype(float) 
sales.standardized_nw = (sales.Price-np.mean(sales.Price))/(np.std(sales.Price))

#Identification and removal of the outliers using Six Sigma
n = np.mean(sales.Price)
s = np.std(sales.Price)
ll = max (0, n-3*s)
ul = n + 3*s
sales_cleaned1 = sales[sales.Price>= ll]
sales_cleaned2 = sales_cleaned1[sales.Price <= ul]
sales_cleaned2

#saving to csv
sales_cleaned2.to_csv("Pareja_sales.csv", index = False)
Pareja_sales = pd.read_csv("Pareja_sales.csv")
Pareja_sales
```

```python
# Part 2: Using the data in Handling Missing Values.csv
# 1. Handle all the missing values. Provide and exlanation for choosing specific method 
#   (delete, impute or keep; if impute, why mean, median or mode?) in handling specific missing values

data = pd.read_csv('Handling_Missing_Values.csv')

#remove data with 2 or more missing values
data1 = data.dropna(thresh = 6)

#filling the sex data with the mode
mode_value=data1['Sex'].mode()
data1['Sex']=data1['Sex'].fillna(mode_value[0])


# using data1.skew(axis = 0, skipna = True) the data is moderately skwed
# use of median to fill the data for income and number of children. 
data1

median_value=data1['Income'].median()
data1['Income']=data1['Income'].fillna(median_value)
data1

median_value1=data1['Children'].median()
data1['Children']=data1['Children'].fillna(median_value1)


#saving to csv
data1.to_csv("Pareja_Handling_Missing_Values.csv", index = False)
Pareja_Handling_Missing_Values = pd.read_csv("Pareja_Handling_Missing_Values.csv")
Pareja_Handling_Missing_Values
```

```python
# Part 3: Application
# 1. Gather Data From Any Source (it would be better if you use your own dta)
# 2. Preprocess the data you gathered.

data = pd.read_excel('Survey.xlsx')
data = data.iloc[0:35,0:12] #remove the empty columns
data = data.dropna(thresh = 10) #use data.shape to check how many rows are omitted

# use data.isnull().sum() to check the column with missing data
# mode is used for categorical data and median is used for numerical data
a = ['Place of Residence', 'Work Background of the Shopper', 'Mode of Transportation', 
     'Frequency of Shopping','Time of Shopping', 'Source of Everyday Food','Day of Shopping',
     'Age of the Person who do the Shopping' ]
for i in a:
    mode_value=data[i].mode()
    data[i]=data[i].fillna(mode_value[0])

b = ['Size of Household','Time Travel (One Way in Minutes)', 
     'Shopping Time (inc the time in queuing in hours)']
for i in b:
    median_value=data[i].median()
    data[i]=data[i].fillna(median_value)

#Identification and removal of the outliers using Six Sigma
#data.describe(include="all") to check the standard deviation
for i in b:
    n = np.mean(data[i])
    s = np.std(data[i])
    ll = max (0, n-3*s)
    ul = n + 3*s
    data = data[data[i]>= ll]
    data = data[data[i] <= ul]

#saving to csv
data.to_csv("Pareja_Survey.csv", index = False)
Pareja_Survey = pd.read_csv("Pareja_Survey.csv")
Pareja_Survey
```
