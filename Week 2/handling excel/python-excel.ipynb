```python
import pandas as pd
from pandas import ExcelWriter

data = pd.read_excel("sales.xlsx", sheet_name='sales data')
filter = data.loc[data['Tax'] > 0.07]

# Determining File
writer = ExcelWriter('empty.xlsx')

# Writing excel file
filter.to_excel(writer,'Sheet1', index=False)
writer.save()
```

```python
import pandas as pd
from pandas import ExcelWriter

# read sakes,xlsx
data = pd.read_excel("df1.xlsx")
data1 = pd.read_excel("df2.xlsx")
data2 = pd.read_excel("df3.xlsx")

x = pd.concat([data,data1,data2])

writer = ExcelWriter("df4.xlsx")
x.to_excel(writer,'Sheet1', index = False)
writer.save()

data3=pd.read_excel('df4.xlsx')
data3
```
