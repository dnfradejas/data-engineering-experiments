# Set 1
Print six scatterplots (1 per district) showing the number of crimes (y-axis) per day (x-axis) for entries dated from January 1 to 10 in each specific district.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
crime_data = pd.read_csv('crime.csv')
```

```python
# convert columns to their specific data type

#convert date to datetime
crime_data["cdatetime"] = pd.to_datetime(crime_data.cdatetime)

#convert district to numeric
crime_data["district"] = pd.to_numeric(crime_data.district)    
```

# District 1 to 6

```python
# all crimes listed from January 1 to 10 on District 1-6
color = ['darkorange', 'forestgreen', 'slategray', 'brown','yellowgreen','deepskyblue'] # set color palettes

# iterate each district from 1-6
for district in range(1,7):
    crime_month = crime_data[(crime_data.cdatetime.dt.day <= 10) &  # dt.day == 10 means the maximum day of the month
                             (crime_data.cdatetime.dt.month == 1) &  # dt.month == 1 means that the month is january
                             (crime_data.district == district)] # select district 1-6


    x = crime_month.cdatetime.dt.strftime('%b-%d').unique() # extract all unique dates from Jan-1 to Jan-10
    y = [] # declare the values for y-axis

    # compute all the total crimes per day where x is being contained
    for days in x:
        y.append((crime_month[crime_month.cdatetime.dt.strftime('%b-%d') == days]).ucr_ncic_code.value_counts().sum())

        
    # Plot the scatter graph of Crimes for January in District 1
    plt.scatter(x,y, c= color[district-1]) #assign color palette per district no.
    plt.title('Crimes in January in Disrict {0}'.format(district))
    plt.xlabel('Day')
    plt.ylabel('Crimes')
    plt.show()
```
