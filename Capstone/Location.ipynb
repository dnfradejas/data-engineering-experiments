```python
import pandas as pd
```

# Identify different cities

```python
cities = ['Bacolod', 'Baguio', 'Batangas', 'Boracay', 'Cubao', 'Tanay Rizal', 'Cagayan de Oro', 'Cebu' ,
          'Coron', 'Davao', 'Dumaguete', 'Lapu-Lapu City', 'El Nido', 'Makati', 'Quezon City','Pasay', 
          'Muntinlupa', 'Mandaluyong', 'Paranaque', 'Manila','Taguig', 'Tagaytay', 'Bohol', 
          'San Isidro', 'Subic', 'Zambales', 'Pampanga', 'Antipolo City', 'Cavite', 'Laguna', 'San Isidro', 'Palawan']

cities.sort()

hotels = pd.read_csv("Hotel.csv")
restaurant = pd.read_csv("Restaurant.csv")
tourist_spot = pd.read_csv("Tourist_site.csv")
```

# Add columns for category and city for Hotels

```python
hotel_city = []
for location in hotels['location']:
    for city in cities:
        if city in location:
            hotel_city.append(city)
            break
            
hotels.insert(2, 'city', hotel_city)
hotels.insert(3, 'category', 'hotel')
hotels
```

# Add columns for category and city for Restaurant

```python
restaurant_city = []
for location in restaurant['location']:
    for city in cities:
        if city in location:
            restaurant_city.append(city)
            break

restaurant.insert(2, 'city', restaurant_city)
restaurant.insert(3, 'category', 'restaurant')
restaurant
```

```python
tourist_spot_city = []
for location in tourist_spot['location']:
    for city in cities:
        if city in location:
            tourist_spot_city.append(city)
            break

tourist_spot.insert(2, 'city', tourist_spot_city)
tourist_spot.insert(3, 'category', 'tourist spot')
tourist_spot
```

# Analyze Columns of Hotels, Restaurants, and Tourist Site
1. Rename columns to that everything will have a common column

```python
hotels.rename(columns={'hotel_name':'name', 'price_per_night':'price_range', 'customer_reviews': 'review'}, inplace=True)
hotels.columns
```

```python
restaurant.rename(columns={'restaurant':'name'}, inplace=True)
restaurant.columns
```

```python
tourist_spot.rename(columns={'tourist_spot':'name', 'entrance_fee':'price_range'}, inplace=True)
tourist_spot.columns
```

```python
combine = pd.concat([hotels[['name', 'location', 'city', 'category', 'price_range','rating', 'review']], 
                    restaurant[['name', 'location', 'city', 'category', 'price_range','rating', 'review']],
                    tourist_spot[['name', 'location', 'city', 'category', 'price_range','rating', 'review']]])
```

# Sort Columns so that there will be a common column

```python
combined_sort = combine.sort_values(by=['city'])
combined_sort = combined_sort.reset_index(drop=True)
combined_sort
```

# Save values to csv file

```python
import pandas as pd

# save to csv
combined_sort.to_csv("Location.csv", index=False)
pd.read_csv("Location.csv")
```
