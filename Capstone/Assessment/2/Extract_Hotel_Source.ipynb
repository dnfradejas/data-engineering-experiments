```python
#import important libraries

import numpy as np
import pandas as pd
import requests
import bs4
from bs4 import BeautifulSoup
import urllib.request
import re #regular expression
import lxml
import json
from time import sleep
from random import randint
```

```python
#get data from Trip Advisor

url = 'https://www.tripadvisor.in/Hotels-g298449-Metro_Manila_Luzon-Hotels.html'

#connect to website
page = urllib.request.urlopen(url) 

#convert to beautiful soup object
soup = BeautifulSoup(page, 'lxml')
print(soup)
```

```python
#create a list where the link of each hotel pages from TripAdvisor will be stored

links = []
```

```python
#find all links of each hotel pages from Trip Advisor and store to the links list

for review in soup.findAll('a',{'class':'review_count'}):
  a = review['href']
  a = 'https://www.tripadvisor.in'+ a
  a = a[:(a.find('Reviews')+7)] + '-or{}' + a[(a.find('Reviews')+7):]
  links.append(a)
```

```python
#check contents of links
links
```

```python
#create another dataframe which will contain the segregated hotel data for reviews per hotel

columns2 = ['hotel_name','review_1','review_2','review_3','review_4','review_5',
            'review_6','review_7','review_8','review_9','review_10',
            'review_11','review_12','review_13','review_14','review_15']
df_reviews = pd.DataFrame(columns = columns2, index = range(0,len(links)))
df_reviews
```

```python
#This section extracts the hotel reviews from the hotel pages of Trip Advisor. Note that each page only contains 5 hotel reviews and 
#need to click next page to view the succeeding reviews.

#create a list which will contain numbers that will be used to extract reviews from page 1 to page 3 of the hotel pages
d = ['','5','10']

l = 0

#using for loop, extract reviews fron each hotel pages of Travel Advisor
for link in links:
    reviews = []
    for i in range(1,len(d)+1):
        
        #replace the {} with the values of d. " ",5,10 refers to page 1,2,3 of the hotel page respectively
        link2 = link.replace("{}",str(d[i-1]))
        soup2 = urllib.request.urlopen(link2)
        sleep(randint(1,5))
        bsobj2 = BeautifulSoup(soup2,'lxml')
        if i == 1:
            
            #extract the headers, which contains the name of the hotel and append to reviews
            for s in bsobj2.find_all('h1',{'class':'_1mTlpMC3'}):
                reviews.append(s.getText().split('\n')[0])

        #extract reviews and append to reviews
        for r in bsobj2.findAll('q'):
            reviews.append(r.span.text.strip())
        
    for j in range(0,len(reviews)):
        if j < len(columns2):
            df_reviews.iloc[l,j] = reviews[j]
        
    if l < len(links)-1:
        l = l +1
```

```python
#check content
df_reviews
```

```python
#create another dataframe which will contain the segregated hotel data for reviewers per hotel

columns3 = ['hotel_name','reviewer_1','reviewer_2','reviewer_3','reviewer_4','reviewer_5',
            'reviewer_6','reviewer_7','reviewer_8','reviewer_9','reviewer_10','reviewer_11','reviewer_12','reviewer_13','reviewer_14','reviewer_15']
df_reviewer = pd.DataFrame(columns = columns3, index = range(0,len(links)))
df_reviewer
```

```python
#This section extracts the hotel reviewers from the hotel pages of Trip Advisor. Note that each page only contains 5 hotel reviews and 
#need to click next page to view the succeeding reviews.

#create a list which will contain numbers that will be used to extract reviews from page 1 to page 3 of the hotel pages
d = ['','5','10']

l = 0

#using for loop, extract reviews fron each hotel pages of Travel Advisor
for link in links:
    reviewer = []
    for i in range(1,len(d)+1):
        
        #replace the {} with the values of d. " ",5,10 refers to page 1,2,3 of the hotel page respectively
        link2 = link.replace("{}",str(d[i-1]))
        soup2 = urllib.request.urlopen(link2)
        sleep(randint(1,5))
        bsobj2 = BeautifulSoup(soup2,'lxml')
        if i == 1:
            
            #extract the headers, which contains the name of the hotel and append to reviews
            for s in bsobj2.find_all('h1',{'class':'_1mTlpMC3'}):
                reviewer.append(s.getText().split('\n')[0])

        #extract reviewers and append to reviews
        for u in bsobj2.find_all('a',{'class':'ui_header_link _1r_My98y'}):
            reviewer.append(u.getText().split('\n')[0])
        
    for j in range(0,len(reviewer)):
        if j < len(columns3):
            df_reviewer.iloc[l,j] = reviewer[j]
        
    if l < len(links)-1:
        l = l +1
```

```python
#check content
df_reviewer
```

```python
#create another dataframe which will contain the segregated hotel data for descriptions per hotel

columns4 = ['hotel_name','about']
df_about = pd.DataFrame(columns = columns4, index = range(0,len(links)))
df_about
```

```python
#This section extracts the hotel descriptions from the hotel pages of Trip Advisor. Note that each page only contains 5 hotel reviews and 
#need to click next page to view the succeeding reviews.

#create a list which will contain numbers that will be used to extract reviews from page 1 to page 3 of the hotel pages

l = 0

#using for loop, extract reviews fron each hotel pages of Travel Advisor
for link in links:
    about = []
        
    #replace the {} with the values of d. " ",5,10 refers to page 1,2,3 of the hotel page respectively
    link2 = link.replace("{}",'')
    soup2 = urllib.request.urlopen(link2)
    sleep(randint(1,5))
    bsobj2 = BeautifulSoup(soup2,'lxml')
            
    #extract the headers, which contains the name of the hotel and append to reviews
    for s in bsobj2.find_all('h1',{'class':'_1mTlpMC3'}):
        about.append(s.getText().split('\n')[0])

    #extract reviewers and append to reviews
    for s in bsobj2.find_all('div',{'class':'cPQsENeY'}):#{'class':'cPQsENeY'}):
        about.append(s.getText().split('\n')[0])
        
    for j in range(0,len(about)):
        if j < len(columns4):
            df_about.iloc[l,j] = about[j]
    
    if (df_about.iloc[l,len(columns4)-1]) == (df_reviews.iloc[l,1]):
        df_about.iloc[l,len(columns4)-1] = ''
        
    if l < len(links)-1:
        l = l +1
```

```python
#check content
df_about = df_about.fillna('No Data')
df_about
```

```python
#create a dataframe which will contain the amenities of each hotel

columns5 = ['hotel_name','amenities']
df_amenities = pd.DataFrame(columns = columns5, index = range(0,len(links)))
df_amenities
```

```python
#create list for hotel name and amenities

hotel_name = []
amenities = []

i = 0
j = 0

#for each link, extract the amenities data

for link in links:
    link3 = link.replace('{}','')
    soup3 = urllib.request.urlopen(link3)
    sleep(randint(1,5))
    bsobj3 = BeautifulSoup(soup3,'lxml')
    l = 1
    
    #extract header which contains the name of the hotel
    for s in bsobj3.find_all('h1'):
        hotel_name.append(s.getText().split('\n')[0])
        if l == 1:
            df_amenities.iloc[i,0] = s.getText().split('\n')[0]
    k = j
    
    #extract all the amenities data and store to df_amenities
    for t in bsobj3.findAll('div',{'class':'_2rdvbNSg'}):
        amenities.append(str(t.getText()))
        j = j + 1
        df_amenities.iloc[i,1] = amenities[k:j+1]
    i = i +1
```

```python
#check content
df_amenities = df_amenities.fillna('No Data')
df_amenities
```

```python
#create a dataframe which will contain the price of each hotel

columns6 = ['hotel_name','price']
df_price = pd.DataFrame(columns = columns6, index = range(0,len(links)))
df_price
```

```python
#This section extracts the hotel prices from the hotel pages of Trip Advisor. 


l = 0

#using for loop, extract prices fron each hotel pages of Travel Advisor
for link in links:
    price = []

    link2 = link.replace("{}",'')
    soup2 = urllib.request.urlopen(link2)
    sleep(randint(1,5))
    bsobj2 = BeautifulSoup(soup2,'lxml')
            
    #extract the headers, which contains the name of the hotel and append to reviews
    for s in bsobj2.find_all('h1',{'class':'_1mTlpMC3'}):
        price.append(s.getText().split('\n')[0])

    #extract prices
    for s in bsobj2.findAll('div',{'class':'_36QMXqQj'}):#{'class':'cPQsENeY'}):
        price.append(s.getText().split('\n')[0])
        
    for j in range(0,len(price)):
        if j < len(columns6):
            if j == len(columns6)-1:
                price[j] = price[j].replace('\xa0','')
            df_price.iloc[l,j] = price[j]
        
    if l < len(links)-1:
        l = l +1
```

```python
#check content
df_price = df_price.fillna('No Data')
df_price
```

```python
#create df_rooms dataframe which will contain the data for the number of rooms of each hotel

columns7 = ['hotel_name','no_of_rooms']
df_rooms = pd.DataFrame(columns = columns7, index = range(0,len(links)))

l = 0

#using for loop, extract the number of hotel rooms fron each hotel pages of Travel Advisor
for link in links:
    rooms = []
    rooms_2 = []
    
    link2 = link.replace("{}",'')
    soup2 = urllib.request.urlopen(link2)
    sleep(randint(1,5))
    bsobj2 = BeautifulSoup(soup2,'lxml')
            
    #extract the headers, which contains the name of the hotel and append to reviews
    for s in bsobj2.find_all('h1',{'class':'_1mTlpMC3'}):
        rooms.append(s.getText().split('\n')[0])

    #extract hotel room numbers
    for t in bsobj2.findAll('div',{'class':'_1NHwuRzF'}):
        rooms.append(t.getText().split('\n')[0])
        
    for j in range(0,len(rooms)):
        if j == 0:
            rooms_2.append(rooms[j])
        if j > 0:
            if rooms[j].isdigit() == True:
                rooms_2.append(rooms[j])
                
    for k in range(0,len(rooms_2)):
        if k < len(columns7):
            df_rooms.iloc[l,k] = rooms_2[k]
    if l < len(links)-1:
        l = l +1
```

```python
#check content
df_rooms = df_rooms.fillna('No Data')
df_rooms
```

```python
#create dataframe for hotel addresses
columns8 = ['hotel_name','address']
df_address = pd.DataFrame(columns = columns8, index = range(0,len(links)))
```

```python
#This section extracts the hotel addresses from each hotel page in Travel Advisor
l = 0

#using for loop, extract reviews fron each hotel pages of Travel Advisor
for link in links:
    address = []
    address_2 = []

    link2 = link.replace("{}",'')
    soup2 = urllib.request.urlopen(link2)
    sleep(randint(1,5))
    bsobj2 = BeautifulSoup(soup2,'lxml')
            
    #extract the headers, which contains the name of the hotel and append to reviews
    for s in bsobj2.find_all('h1',{'class':'_1mTlpMC3'}):
        address.append(s.getText().split('\n')[0])

    #extract hotel addresses
    for t in bsobj2.findAll('span','_3ErVArsu jke2_wbp'):#{'class':'cPQsENeY'}):
        address.append(t.getText().split('\n')[0])
        
    for j in range(0,len(address)):
        if j == 0:
            address_2.append(address[j])
        if j == 1:
            address_2.append(address[j])
                
    for k in range(0,len(address_2)):
        if k < len(columns8):
            df_address.iloc[l,k] = address_2[k]
    if l < len(links)-1:
        l = l +1
```

```python
#check content
df_address
```

```python
#merge all dataframes
df_hotel = df_reviews.merge(df_reviewer, on = 'hotel_name', how = 'left')
df_hotel = df_hotel.merge(df_about, on = 'hotel_name', how = 'left')
df_hotel = df_hotel.merge(df_amenities, on = 'hotel_name', how = 'left')
df_hotel = df_hotel.merge(df_price,on = 'hotel_name', how = 'left')
df_hotel = df_hotel.merge(df_rooms,on = 'hotel_name', how = 'left')
df_hotel = df_hotel.merge(df_address,on = 'hotel_name', how = 'left')
```

```python
#reorder columns
df_hotel = df_hotel[['hotel_name','about','address','price','no_of_rooms','amenities','review_1','reviewer_1','review_2','reviewer_2',
                    'review_3','reviewer_3','review_4','reviewer_4','review_5','reviewer_5','review_6','reviewer_6','review_7',
                    'reviewer_7','review_8','reviewer_8','review_9','reviewer_9','review_10','reviewer_10','review_11',
                    'reviewer_11','review_12','reviewer_12','review_13','reviewer_13','review_14','reviewer_14','review_15',
                    'reviewer_15']]
df_hotel.info()
```
