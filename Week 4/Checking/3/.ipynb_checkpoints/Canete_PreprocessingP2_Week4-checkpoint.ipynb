```python
#Import libraries 

import pandas as pd
import numpy as np
```

```python
#read csv file and save it to sales variable

sales = pd.read_csv('D:/Personal Files/Data Scientist Track - Project Sparta/Python for Data Engineering/sales.csv')
```

### Data Preprocessing

```python
#check dataframe for presence of null values

sales.isnull().sum()
```

```python
#Fill Missing Values for Transaction_date

sales['Transaction_date'] = sales['Transaction_date'].fillna('No TimeStamp')
```

```python
#Fill Missing Values for Product
mode_Product = sales['Product'].mode().iloc[0]
sales['Product'] = sales['Product'].fillna(mode_Product)
```

```python
#Fill Missing Values for Price

sales['Price'] = sales.Price.str.replace(',','')
sales['Price'] = sales.Price.astype(float)
mean_value = sales['Price'].mean()
sales['Price']=sales['Price'].fillna(mean_value)
```

```python
sales.Price
```

```python
#Fill Missing Values for Payment Type
mode_Payment = sales['Payment_Type'].mode().iloc[0]
sales['Payment_Type'] = sales['Payment_Type'].fillna(mode_Payment)
```

```python
#Fill Missing Values for City
sales['City'] = sales['City'].fillna('No Data')
```

```python
#Fill Missing Values for State
sales['State'] = sales['State'].fillna('No Data')
```

```python
#Fill Missing Values for Account Created

sales['Account_Created'] = sales['Account_Created'].fillna('No TimeStamp')
```

```python
#Fill Missing Values for Account Created

sales['Last_Login'] = sales['Last_Login'].fillna('No TimeStamp')
```

```python
#Fill Missing Values for Latitude

mean_latitude = sales['Latitude'].mean()
sales['Latitude']=sales['Latitude'].fillna(mean_latitude)
```

```python
#Fill Missing Values for Longitude

mean_longitude = sales['Longitude'].mean()
sales['Longitude']=sales['Longitude'].fillna(mean_longitude)
```

```python
sales.isnull().sum()
```

```python
sales.head()
```

### Standardized Price Column

```python
sales.insert(12, 'Standardized_Price', 0)
```

```python
sales['Standardized_Price'] = (sales.Price - np.mean(sales.Price))/ (np.std(sales.Price))
```

```python
sales.describe(include = 'all')
```

```python
sales
```

### Part 2 Handle missing Values using Handling_Missing_values.csv

```python
handle = pd.read_csv('D:/Personal Files/Data Scientist Track - Project Sparta/Python for Data Engineering/Handling_Missing_Values.csv')
```

```python
handle
```

```python
handle.isnull().sum()
```

```python
# Drop/ Delete a row that has 5 values that have not-NAN
handle = handle.dropna(thresh=5)
```

```python
handle.describe(include = 'all')
```

```python
handle
```

```python
handle.isnull().sum()
```

```python
# Since Income is a numerical we can impute the NAN values with MEDIAN

income_mean = handle['Income'].mean()
handle['Income'] = handle['Income'].fillna(income_mean)
```

```python
handle
```

```python
# Since Sex is a Categorical we can impute the NAN values with MODE

handle['Sex']= handle['Sex'].fillna(handle['Sex'].mode()[0])
```

```python
handle
```

```python
# Since No of Children is a numerical but cannot be in terms of floating number/decimal, we can impute the NAN values with MODE

handle['Children']= handle['Children'].fillna(handle['Children'].mode()[0])
```

```python
handle
```

```python
handle.to_csv('D:/Personal Files/Data Scientist Track - Project Sparta/Python for Data Engineering/canete_handle_missing_file.csv', index = False)
```
