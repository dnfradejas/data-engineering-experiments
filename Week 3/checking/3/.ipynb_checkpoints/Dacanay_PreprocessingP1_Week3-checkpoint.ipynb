```python
sales_data_nodups.reset_index()
print(sales_data_nodups.Price)
print("\nData Type is Object. check if all data is numeric\n\n")
sales_data_nodups.Price = pd.to_numeric(sales_data_nodups.Price) 
##Error is part of my presented answer
```

**Finding: Entry with String 13,000 is not convertible. Remove Entry**

```python
sales_data_nodups[sales_data_nodups['Price']=='13,000']
```

**Findings: Use Index 560**

```python

sales_data_nodups2 = sales_data_nodups.drop([560]) #drops the 13K error
sales_data_nodups2.Price = pd.to_numeric(sales_data_nodups2.Price)
sales_data_nodups2.Price = sales_data_nodups2.Price/1000 #Price Convertion to Thousand Units
sales_data_nodups2.head()
```

## 4. Select the name, city, state, and country data of product 1 bought by customers from United States and Paid Thru Amex or Diners. Store the selected data in empty.csv file. 

```python
prod1_to_file = sales_data_nodups2[(sales_data_nodups2['Country']=='United States') &\
                   (sales_data_nodups2['Product']=='Product1') &\
                  ((sales_data_nodups2['Payment_Type']=='Amex') |\
                   (sales_data_nodups2['Payment_Type']=='Diners'))]\
                  [['Name','City','State','Country']]
prod1_to_file.to_csv('empty.csv', index = False)
prod1 = pd.read_csv('empty.csv')
prod1 #test if specified data was written to file
```
