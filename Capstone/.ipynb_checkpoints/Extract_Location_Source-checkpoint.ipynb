# Language Used
1. Source: https://data.humdata.org/dataset/philippines-languages

```python
language_main = pd.read_csv("ph_lang_admin2_v01.csv")
region_language = language_main[['admin2_pcode','admin1_name','main_language']] 
region_language.admin2_pcode
```

```python
def get_language_index(location):
    location = location.capitalize()
    if location in region_language.admin2_pcode.values: # if location is existing in Municipal Level
        index = region_language.index[region_language['admin2_pcode'] == location][0]
    else:
        index = 64
    return index
```

```python
locations = ['ANTIPOLO CITY',
 'BACOLOD',
 'BAGUIO',
 'BATANGAS',
 'BOHOL',
 'BORACAY',
 'CAGAYAN DE ORO',
 'CAVITE',
 'CEBU',
 'CORON',
 'CUBAO',
 'DAVAO',
 'DUMAGUETE',
 'EL NIDO',
 'LAGUNA',
 'LAPU-LAPU CITY',
 'MAKATI',
 'MANDALUYONG',
 'MANILA',
 'MUNTINLUPA',
 'PALAWAN',
 'PAMPANGA',
 'PASAY',
 'QUEZON CITY',
 'SAN ISIDRO',
 'SUBIC',
 'TAGAYTAY',
 'TAGUIG',
 'TANAY RIZAL',
 'ZAMBALES']
```

```python
for city in locations:
    main_language.append(region_language.main_language[get_language_index(city)])
    secondary_language.append('Tagalog')
```
