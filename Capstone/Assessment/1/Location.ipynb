Tourist data per Location

```python
import pandas as pd

hotel = pd.read_csv('Hotel.csv')
restaurant = pd.read_csv('Restaurant.csv')
tourist = pd.read_csv('Tourist_site.csv')
```

```python
print (hotel)
print ("---------end------------")
print (restaurant)
print ("---------end------------")
print (tourist)
print ("---------end------------")
```

```python
tourism = pd.concat([hotel,restaurant,tourist])
tourism
```

```python
#List unique values in the Province column
tourism.Province.unique()
```

```python
# Now we that we have unique names from the province column, we will now create a csv file for each Location

#['Busuanga Island', 'NCR', 'Calabarzon', 'Mindanao', 'Cebu',
#       'Baguio', 'Ilocos', 'Palawan', 'Legaspi', 'Ifugao', 'Siargao',
#       'Bohol', 'Boracay', 'Mountain Province', 'Cavite', 'Bataan',
#       'Negros Occidental', 'Laguna', 'Sorsogon', 'Romblon'], dtype=object)

#Busuanga Island

BusuangaIsland = tourism[tourism.Province == 'Busuanga Island']
print (BusuangaIsland)
```
